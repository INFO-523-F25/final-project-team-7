# Week 03 — Statistical Detectors (Z-score, Elliptic Envelope, Mahalanobis)
Resampled subset `{normal, teardrop}`. Numeric-only features for multivariate methods.
Predictions use anomaly=1 and normal=0.

```python

import numpy as np, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
from src.utils import DATA_RAW, FIGURES, ensure_dir, RANDOM_STATE
from src.io import load_nsl_kdd_raw
from src.models import elliptic_envelope_predict, mahalanobis_predict

FIG_DIR = ensure_dir(FIGURES)
RAW_FILE = DATA_RAW / "NSL-KDD.raw"
print("Raw file:", RAW_FILE)
print("Figures dir:", FIG_DIR)
```

```python

# Build subset {normal, teardrop}. Fallback to rarest attack if teardrop absent.
df = load_nsl_kdd_raw(RAW_FILE).copy()
df["label"] = df["label"].astype(str).str.rstrip(".")
labels = df["label"].unique().tolist()

target_attack = "teardrop" if "teardrop" in labels else None
if target_attack is None:
    vc = df["label"].value_counts(); vc = vc[vc.index != "normal"]
    if len(vc) == 0:
        raise ValueError("No attack labels present in this file.")
    target_attack = vc.sort_values(ascending=True).index[0]
    print(f"⚠️ Using rarest available attack as anomaly class: '{target_attack}'.")
else:
    print("Using 'teardrop' as anomaly class.")

subset = df[df["label"].isin(["normal", target_attack])].copy()
subset["y"] = (subset["label"] == target_attack).astype(int)
categorical = ["protocol_type","service","flag","label"]
num_cols = [c for c in subset.columns if c not in categorical + ["y"]]
X_num = subset[num_cols]
y_bin = subset["y"].to_numpy(int)

print("Subset counts:", subset["y"].value_counts().to_dict(), "| Numeric features:", len(num_cols))
```

## Step 1 — Univariate Z-score on `wrong_fragment`

```python

assert "wrong_fragment" in subset.columns, "Missing wrong_fragment feature."
wf = pd.to_numeric(subset["wrong_fragment"], errors="coerce").fillna(0.0)
mu = float(wf.mean()); sd = float(wf.std(ddof=0)) if wf.std(ddof=0) > 0 else 1e-9
z = (wf.to_numpy() - mu) / sd
y_pred_z = ((z > 2) | (z < -2)).astype(int)

cm_z = confusion_matrix(y_bin, y_pred_z, labels=[0,1])
print("Confusion matrix [[TN, FP],[FN, TP]]:\n", cm_z)
print("Classification report:\n", classification_report(y_bin, y_pred_z, target_names=["normal", target_attack], zero_division=0))

plt.figure(figsize=(6,5))
sns.heatmap(cm_z, annot=True, fmt="d", cmap="YlGnBu", cbar=True)
plt.xlabel("Predicted label"); plt.ylabel("True label")
plt.title("Week 03 — Z-score on wrong_fragment (±2σ)")
out_z = FIG_DIR / "week3_confusion_zscore.png"
plt.tight_layout(); plt.savefig(out_z, dpi=150); plt.show()
print("Saved confusion matrix to:", out_z)
```

## Step 3 — Elliptic Envelope (numeric-only)


## DBSCAN (Density-Based) — Week 03

This section adds a density-based anomaly detector consistent with Weeks 01–02 setup:
- Uses **existing modules** for project constants (`src.utils`) and raw access (`src.io` already used elsewhere).
- Consumes the processed test split from Week 01 (`data/processed/X_test.npy`, `y_test.npy`).
- Labels DBSCAN noise (`-1`) as **anomaly = 1** and others as **0**.
- Runs a compact parameter grid and selects the configuration with the best **F1** on the test set.
- Saves a confusion matrix figure to the standard `FIGURES` directory.


```python
# DBSCAN (low-memory, module-aligned)
# In this task we run a density-based anomaly detector on the processed test split from Week 01.

from pathlib import Path
import json, gc
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn.metrics import (
    confusion_matrix, ConfusionMatrixDisplay, classification_report,
    precision_recall_fscore_support
)
from sklearn.decomposition import PCA, TruncatedSVD
from sklearn.preprocessing import StandardScaler

# --- Project paths from your modules ---
FIG_DIR = ensure_dir(FIGURES)

# Prefer a constant from utils if you have it; else fall back to your absolute path, then repo-relative
try:
    from src.utils import DATA_PROCESSED
    PROCESSED_DIR = Path(DATA_PROCESSED)
except Exception:
    for cand in [Path(r"C:\Users\mehra\Final_Project\data\processed"), Path("data/processed")]:
        if (cand / "X_test.npy").exists():
            PROCESSED_DIR = cand
            break
    else:
        raise FileNotFoundError("Processed data directory not found.")

print("Processed dir:", PROCESSED_DIR)
print("Figures dir:", FIG_DIR)

# --- Robust & memory-friendly loader (handles object arrays and sparse payloads) ---
def load_npy_memsafe(path: Path):
    try:
        return np.load(path, mmap_mode="r")  # no full copy
    except ValueError as e:
        # Object array (e.g., saved sparse) → allow_pickle + unwrap
        arr = np.load(path, allow_pickle=True)
        if getattr(arr, "dtype", None) == object and arr.size == 1:
            arr = arr.item()
        try:
            import scipy.sparse as sp
            if sp.issparse(arr):
                return arr  # keep sparse; don't densify here
        except Exception:
            pass
        return np.asarray(arr)

# --- Normalize labels to {0,1} (idempotent) ---
def to_binary_anomaly(y):
    import numpy as _np
    arr = _np.asarray(y)
    try:
        uniq = _np.unique(arr)
        if set(uniq.tolist()).issubset({0,1}) or set(uniq.astype(int).tolist()).issubset({0,1}):
            return arr.astype(int, copy=False)
    except Exception:
        pass
    if arr.dtype.kind in {"U","S","O"}:
        lowered = _np.char.lower(arr.astype(str))
        is_anom = _np.ones_like(lowered, dtype=int)
        is_anom[(lowered == "normal") | (lowered == "0")] = 0
        return is_anom
    try:
        return (arr.astype(float) != 0.0).astype(int)
    except Exception:
        return (_np.asarray(arr).astype(str) != "normal").astype(int)

# --- Load processed split (no regeneration) ---
X_test_raw = load_npy_memsafe(PROCESSED_DIR / "X_test.npy")
y_test = to_binary_anomaly(np.load(PROCESSED_DIR / "y_test.npy"))

print("X_test_raw type:", type(X_test_raw))

# --- Prepare matrix for DBSCAN: cap rows, reduce dims, float32 ---
MAX_ROWS   = 30000  # lower if memory is tight; raise if you have headroom
TARGET_DIM = 30     # 20–50 is a good range
rng = np.random.default_rng(RANDOM_STATE)

def stratified_cap_indices(y, cap):
    y = np.asarray(y)
    n = y.shape[0]
    if n <= cap:
        return np.arange(n)
    uniq, counts = np.unique(y, return_counts=True)
    quotas = np.maximum((counts / n * cap).astype(int), 1)
    idxs = []
    for cls, k in zip(uniq, quotas):
        cls_idx = np.flatnonzero(y == cls)
        take = min(k, cls_idx.size)
        idxs.append(rng.choice(cls_idx, size=take, replace=False))
    return np.concatenate(idxs)

def prepare_for_dbscan(X, y):
    sel = stratified_cap_indices(y, MAX_ROWS)
    y_sel = y[sel]
    # sparse-aware path
    try:
        import scipy.sparse as sp
        is_sparse = sp.issparse(X)
    except Exception:
        is_sparse = False
    if is_sparse:
        X_sel = X[sel]
        n_comp = min(TARGET_DIM, max(2, X_sel.shape[1]-1))
        Z = TruncatedSVD(n_components=n_comp, random_state=RANDOM_STATE).fit_transform(X_sel).astype(np.float32, copy=False)
    else:
        X_sel = np.asarray(X)[sel]
        if X_sel.ndim == 1:
            X_sel = X_sel.reshape(-1, 1)
        X_sel = X_sel.astype(np.float32, copy=False)
        if X_sel.shape[1] > TARGET_DIM:
            Z = PCA(n_components=TARGET_DIM, svd_solver="randomized", random_state=RANDOM_STATE).fit_transform(X_sel).astype(np.float32, copy=False)
        else:
            Z = X_sel
    Z = StandardScaler().fit_transform(Z).astype(np.float32, copy=False)
    return Z, y_sel, sel

X_db, y_db, sel_idx = prepare_for_dbscan(X_test_raw, y_test)
print("DBSCAN input:", X_db.shape, X_db.dtype)

# --- DBSCAN grid (kept modest) ---
#eps_grid        = [0.5, 0.8, 1.0, 1.2]
#min_samples_grid= [5, 10, 20, 35]
eps_grid        = [0.2, 0.8]
min_samples_grid= [5, 1000]

best = {"f1": -1.0}
for eps in eps_grid:
    for ms in min_samples_grid:
        db = DBSCAN(eps=eps, min_samples=ms, algorithm="brute", n_jobs=-1)
        labels = db.fit_predict(X_db)
        y_pred = (labels == -1).astype(int)  # noise → anomaly
        prec, rec, f1, _ = precision_recall_fscore_support(y_db, y_pred, average="binary", zero_division=0)
        if f1 > best["f1"]:
            best = {
                "f1": float(f1), "eps": float(eps), "min_samples": int(ms),
                "labels": labels, "metrics": {"precision": float(prec), "recall": float(rec), "f1": float(f1)}
            }

# --- Evaluate on the (possibly capped) set & save artifacts ---
y_pred_final = (best["labels"] == -1).astype(int)
cm = confusion_matrix(y_db, y_pred_final, labels=[0,1])
report = classification_report(y_db, y_pred_final, digits=4)

fig, ax = plt.subplots(figsize=(5,5))
ConfusionMatrixDisplay(cm, display_labels=["Normal","Anomaly"]).plot(ax=ax, colorbar=False)
ax.set_title(f"DBSCAN (low-mem)  eps={best['eps']}, min_samples={best['min_samples']}  n={X_db.shape[0]}")
(fig := fig).savefig(FIG_DIR / "dbscan_confusion_matrix_lowmem.png", bbox_inches="tight", dpi=150)
plt.close(fig)

with open(FIG_DIR / "dbscan_params_lowmem.json", "w") as f:
    json.dump({"eps": best["eps"], "min_samples": best["min_samples"],
               **(best.get("metrics") or {}), "n_rows": int(X_db.shape[0]), "n_dim": int(X_db.shape[1])}, f, indent=2)

print("DBSCAN (low-mem) — best params:", {"eps": best["eps"], "min_samples": best["min_samples"]})
print("DBSCAN (low-mem) — metrics:", best.get("metrics"))
print("Confusion matrix:\n", cm)
print("Classification report:\n", report)

del X_db; gc.collect()
```
