```python
# Environment setup: project root on path, imports, and global seed.
import os, sys
sys.path.append(os.path.abspath(".."))  # from notebooks/ to project root

from src.utils import set_global_seed, Paths

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from pathlib import Path

set_global_seed(42)
paths = Paths().ensure()
print("Imports OK. Using paths:", paths)
```

## 1. Download KDDCup 10% data and construct DataFrame

```python
# Download KDDCup 10% training data + names, cache under paths.data_raw, and build DataFrame.
from requests import get
import gzip

train_data_page = "http://kdd.ics.uci.edu/databases/kddcup99/kddcup.data_10_percent.gz"
labels_page = "http://kdd.ics.uci.edu/databases/kddcup99/kddcup.names"

raw_dir = paths.data_raw
raw_dir.mkdir(parents=True, exist_ok=True)

gz_path = raw_dir / "kddcup.data_10_percent.gz"
txt_path = raw_dir / "kddcup.data_10_percent"

# Download gz if needed
if not gz_path.exists():
    print("Downloading training data from UCI...")
    r = get(train_data_page)
    r.raise_for_status()
    gz_path.write_bytes(r.content)
else:
    print("Found existing gz file at", gz_path)

# Decompress to plain text if needed
if not txt_path.exists():
    print("Decompressing gz to text...")
    with gzip.open(gz_path, "rb") as f_in, open(txt_path, "wb") as f_out:
        f_out.write(f_in.read())
else:
    print("Found existing text file at", txt_path)

# Read raw lines
with open(txt_path, "r") as ff:
    lines = [ln.strip().split(",") for ln in ff.readlines() if ln.strip()]

print("Loaded", len(lines), "rows from KDDCup 10% file.")

# Download column labels (names)
print("Downloading names file...")
resp = get(labels_page)
resp.raise_for_status()
names_text = resp.text

labels = [ln.split(",")[0].split(":") for ln in names_text.split("\n")]
labels = [l for l in labels if l[0]]  # drop empties
final_labels = labels[1:]             # skip first meta line

col_names = [l[0] for l in final_labels] + ["target"]
data = pd.DataFrame(lines, columns=col_names)

# Cast continuous features to float
for idx, l in enumerate(final_labels):
    if l[1].strip() == "continuous.":
        data.iloc[:, idx] = data.iloc[:, idx].astype(float)

print("Data shape:", data.shape)
data.head()
```

```python
# Check target distribution (attack types vs normal).
data['target'].value_counts()
```

## 2. Z-Score baseline on `wrong_fragment` (teardrop vs normal only)

```python
# Filter to normal vs teardrop only and build binary label.
data_resampled = data.loc[data["target"].isin(["normal.", "teardrop."])].copy()

def map_label(target):
    return 0 if target == "normal." else 1

data_resampled["Label"] = data_resampled["target"].apply(map_label)
data_resampled[["target", "Label"]].value_counts()
```

```python
# Compute Z-score for 'wrong_fragment' and density plot.
from sklearn.metrics import confusion_matrix

mu = data_resampled["wrong_fragment"].mean()
sigma = data_resampled["wrong_fragment"].std(ddof=0)
data_resampled["Z"] = (data_resampled["wrong_fragment"] - mu) / sigma

ax = data_resampled["Z"].plot(kind="density")
ax.set_title("Z-score of wrong_fragment (teardrop vs normal)")
ax.set_xlabel("Z-score")
fig = ax.get_figure()
fig.savefig(paths.figs / "week04_zscore_density.png", dpi=150, bbox_inches="tight")
plt.close(fig)

data_resampled.head()
```

```python
# Map Z-score to predicted anomaly label (|z| > 2 => anomaly) and show confusion matrix.
import seaborn as sns

def map_z_to_label(z):
    return 1 if (z > 2 or z < -2) else 0

data_resampled["Z_Label"] = data_resampled["Z"].apply(map_z_to_label)

conf = confusion_matrix(data_resampled["Label"], data_resampled["Z_Label"])
plt.figure(figsize=(6, 5))
sns.heatmap(conf, annot=True, fmt="d", cmap="YlGnBu")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Z-Score baseline on wrong_fragment (|z|>2 as anomaly)")
plt.show()

conf
```

## 3. Elliptic Envelope

```python
# Fit EllipticEnvelope (robust covariance) on numeric features and evaluate.
from sklearn.covariance import EllipticEnvelope

actual_labels = data_resampled["Label"].to_numpy()
X = data_resampled.drop(columns=["Label", "target", "protocol_type", "service", "flag"])

clf = EllipticEnvelope(contamination=0.1, random_state=42)
clf.fit(X)
predicted_labels = clf.predict(X)  # -1 for outliers, 1 for inliers

predicted_labels_rescored = np.where(predicted_labels == -1, 1, 0)

conf_ell = confusion_matrix(actual_labels, predicted_labels_rescored)
plt.figure(figsize=(6, 5))
sns.heatmap(conf_ell, annot=True, fmt="d", cmap="YlGnBu")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("EllipticEnvelope (contamination=0.1)")
plt.show()

conf_ell
```

## 4. Isolation Forest

```python
# IsolationForest on same feature set and confusion matrix.
from sklearn.ensemble import IsolationForest

actual_labels = data_resampled["Label"].to_numpy()
X = data_resampled.drop(columns=["Label", "target", "protocol_type", "service", "flag"])

iso = IsolationForest(random_state=42, contamination=0.1)
iso.fit(X)
predicted_labels = iso.predict(X)  # -1 outlier, 1 inlier

predicted_labels_rescored = np.where(predicted_labels == -1, 1, 0)

conf_iso = confusion_matrix(actual_labels, predicted_labels_rescored)
plt.figure(figsize=(6, 5))
sns.heatmap(conf_iso, annot=True, fmt="d", cmap="YlGnBu")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Isolation Forest (contamination=0.1)")
plt.show()

conf_iso
```

## 5. Local Outlier Factor (LOF)

```python
# Single LOF run with fixed k, plus parameter sweep over k for accuracy / precision / recall curves.
from sklearn.neighbors import LocalOutlierFactor

actual_labels = data_resampled["Label"].to_numpy()
X = data_resampled.drop(columns=["Label", "target", "protocol_type", "service", "flag"])

# Simple run
k_simple = 5
lof_simple = LocalOutlierFactor(n_neighbors=k_simple, contamination=0.1)
predicted_labels = lof_simple.fit_predict(X)   # -1 outlier, 1 inlier
predicted_labels_rescored = np.where(predicted_labels == -1, 1, 0)

conf_lof_simple = confusion_matrix(actual_labels, predicted_labels_rescored)
plt.figure(figsize=(6, 5))
sns.heatmap(conf_lof_simple, annot=True, fmt="d", cmap="YlGnBu")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title(f"LOF (k={k_simple})")
plt.show()

conf_lof_simple
```

```python
# Parameter sweep for LOF over k (neighbors) to reproduce Chapter 2 style curves.
from sklearn.metrics import confusion_matrix

actual_labels = data_resampled["Label"].to_numpy()
X = data_resampled.drop(columns=["Label", "target", "protocol_type", "service", "flag"])

all_k = []
all_accuracies = []
all_precision = []
all_recall = []

total_num_examples = len(X)
start_k = 100
end_k = 1500
step_k = 100

for k in range(start_k, end_k, step_k):
    print(f"Checking for k = {k}")
    lof = LocalOutlierFactor(n_neighbors=k, contamination=0.1)
    predicted_labels = lof.fit_predict(X)
    predicted_labels_rescored = np.where(predicted_labels == -1, 1, 0)
    conf = confusion_matrix(actual_labels, predicted_labels_rescored)

    accuracy = 100.0 * (conf[0, 0] + conf[1, 1]) / total_num_examples
    precision = 100.0 * conf[1, 1] / (conf[1, 1] + conf[1, 0] + 1)
    recall = 100.0 * conf[1, 1] / (conf[1, 1] + conf[0, 1] + 1)

    all_k.append(k)
    all_accuracies.append(accuracy)
    all_precision.append(precision)
    all_recall.append(recall)

plt.figure(figsize=(7, 5))
plt.plot(all_k, all_accuracies, label="Accuracy")
plt.plot(all_k, all_precision, label="Precision")
plt.plot(all_k, all_recall, label="Recall")
plt.xlabel("k (neighbors)")
plt.ylabel("Metric (%)")
plt.title("LOF metrics vs k")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig(paths.figs / "week04_lof_metrics_vs_k.png", dpi=150)
plt.show()
```

## 6. DBSCAN

```python
# DBSCAN clustering: treat label -1 as anomaly, others as normal.
from sklearn.cluster import DBSCAN

actual_labels = data_resampled["Label"].to_numpy()
X = data_resampled.drop(columns=["Label", "target", "protocol_type", "service", "flag"])

epsilon = 0.2

minimum_samples = 5

dbscan = DBSCAN(eps=epsilon, min_samples=minimum_samples)
predicted_labels = dbscan.fit_predict(X)  # -1 outlier, >=0 cluster

predicted_labels_rescored = np.where(predicted_labels == -1, 1, 0)

conf_dbscan = confusion_matrix(actual_labels, predicted_labels_rescored)
plt.figure(figsize=(6, 5))
sns.heatmap(conf_dbscan, annot=True, fmt="d", cmap="YlGnBu")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title(f"DBSCAN (eps={epsilon}, min_samples={minimum_samples})")
plt.show()

conf_dbscan
```

## 7. One-Class SVM

```python
# One-Class SVM with RBF kernel; label -1 as anomaly.
from sklearn import svm

actual_labels = data_resampled["Label"].to_numpy()
X = data_resampled.drop(columns=["Label", "target", "protocol_type", "service", "flag"])

ocsvm = svm.OneClassSVM(kernel="rbf", gamma="scale", nu=0.1)
ocsvm.fit(X)
predicted_labels = ocsvm.predict(X)  # -1 outlier, 1 inlier

predicted_labels_rescored = np.where(predicted_labels == -1, 1, 0)

conf_ocsvm = confusion_matrix(actual_labels, predicted_labels_rescored)
plt.figure(figsize=(6, 5))
sns.heatmap(conf_ocsvm, annot=True, fmt="d", cmap="YlGnBu")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("One-Class SVM (RBF, nu=0.1)")
plt.show()

conf_ocsvm
```

## 8. Autoencoder

```python
# Autoencoder on the same feature set; anomalies by reconstruction error.
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix

actual_labels = data_resampled["Label"].to_numpy()
X = data_resampled.drop(columns=["Label", "target", "protocol_type", "service", "flag"]).to_numpy().astype("float32")

X_train, X_test, y_train, y_test = train_test_split(
    X, actual_labels, test_size=0.33, random_state=42, stratify=actual_labels
)

input_dim = X_train.shape[1]

input_layer = keras.Input(shape=(input_dim,))
encoded = layers.Dense(30, activation="relu")(input_layer)
encoded = layers.Dense(16, activation="relu")(encoded)
encoded = layers.Dense(8, activation="relu")(encoded)

decoded = layers.Dense(16, activation="relu")(encoded)
decoded = layers.Dense(30, activation="relu")(decoded)
decoded = layers.Dense(input_dim, activation="sigmoid")(decoded)

autoencoder = keras.Model(input_layer, decoded)
autoencoder.compile(optimizer="adam", loss="mse")
autoencoder.summary()
```

```python
# Train autoencoder.
history = autoencoder.fit(
    X_train, X_train,
    epochs=10,
    batch_size=256,
    shuffle=True,
    validation_split=0.1,
    verbose=1
)
```

```python
# Compute reconstruction errors on test set and derive anomaly labels.
predicted = autoencoder.predict(X_test)
errors = np.linalg.norm(X_test - predicted, axis=1)

thresh = np.percentile(errors, 99)
print("Reconstruction error threshold (99th percentile):", thresh)

predicted_labels = (errors > thresh).astype(int)

conf_ae = confusion_matrix(y_test, predicted_labels)
plt.figure(figsize=(6, 5))
sns.heatmap(conf_ae, annot=True, fmt="d", cmap="YlGnBu")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Autoencoder anomaly detection (threshold = 99th percentile error)")
plt.show()

conf_ae
```
