# Week 3 — Elliptic Envelope (Normal vs Teardrop)
Fit on normal-only resampled data; drop categorical features; remap predictions (-1→1 anomaly, 1→0 normal).

```python
# --- Univariate Z-score on `wrong_fragment` for {normal, teardrop}. ---
# This cell computes μ, σ on the entire subset, applies ±2σ, and plots the CM.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
import warnings
warnings.filterwarnings("ignore", message=".*covariance matrix associated.*")


from src.utils import DATA_RAW, FIGURES, ensure_dir
from src.io import load_nsl_kdd_raw

# 1) Load and keep only normal/teardrop. Strip trailing dots to match both variants.
df = load_nsl_kdd_raw(DATA_RAW / "NSL-KDD.raw").copy()
df["label"] = df["label"].astype(str).str.rstrip(".")
need = {"normal", "teardrop"}
missing = need - set(df["label"].unique())
assert not missing, f"Required labels missing: {missing}. Provide a file that includes teardrop."

data_resampled = df[df["label"].isin(["normal", "teardrop"])].copy()
data_resampled["Label"] = (data_resampled["label"] == "teardrop").astype(int)

# 2) Prepare the univariate feature and compute z-scores on the entire subset.
data_resampled["wrong_fragment"] = pd.to_numeric(
    data_resampled["wrong_fragment"], errors="coerce"
).fillna(0.0)
mu = data_resampled["wrong_fragment"].mean()     # Use mean of the whole subset.
sigma = data_resampled["wrong_fragment"].std()   # Use sample std of the whole subset.
sigma = sigma if sigma > 0 else 1e-9             # Avoid division by zero.

data_resampled["Z"] = (data_resampled["wrong_fragment"] - mu) / sigma

# 3) Apply two-sided ±2σ rule and get predictions.
def map_z_to_label(z):
    return 1 if (z > 2 or z < -2) else 0

data_resampled["Predicted_Label"] = data_resampled["Z"].apply(map_z_to_label).astype(int)

# 4) Confusion matrix over the full subset (no train/test split).
cm = confusion_matrix(data_resampled["Label"], data_resampled["Predicted_Label"], labels=[0, 1])

# 5) Plot to match your style.
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt="d", cmap="YlGnBu", cbar=True)
plt.xlabel("Predicted label")
plt.ylabel("True label")
plt.title("Confusion matrix — Z-score on wrong_fragment")
fig_path = ensure_dir(FIGURES) / "week3_confusion_matrix_zscore_wrong_fragment_fullset.png"
plt.tight_layout()
plt.savefig(fig_path, dpi=150)
plt.show()

print("Counts (true):", data_resampled["Label"].value_counts().to_dict())
print("Counts (pred):", data_resampled["Predicted_Label"].value_counts().to_dict())
print("Saved confusion matrix to:", fig_path)
```

```python
# --- Elliptic Envelope on {normal, teardrop}; drop categoricals; fit on full set. ---
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.covariance import EllipticEnvelope
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, classification_report

from src.utils import DATA_RAW, FIGURES, ensure_dir, RANDOM_STATE
from src.io import load_nsl_kdd_raw

# 1) Load and keep only normal + teardrop. Normalize label tokens.
df = load_nsl_kdd_raw(DATA_RAW / "NSL-KDD.raw").copy()
df["label"] = df["label"].astype(str).str.rstrip(".")
subset = df[df["label"].isin(["normal", "teardrop"])].copy()
subset["y"] = (subset["label"] == "teardrop").astype(int)  # normal=0, teardrop=1

# 2) Build numeric feature matrix. Drop categorical features.
categorical = ["protocol_type", "service", "flag", "label"]
num_cols = [c for c in subset.columns if c not in categorical + ["y"]]
X = subset[num_cols].apply(pd.to_numeric, errors="coerce").fillna(0.0).to_numpy(float)
y = subset["y"].to_numpy(int)

# 3) Standardize features.
scaler = StandardScaler(with_mean=True, with_std=True)
Xz = scaler.fit_transform(X)

# 4) Fit Elliptic Envelope on the full resampled set. Use fixed contamination ~10%.
contamination = 0.10  # This yields ~10% predicted anomalies like the shown figure.
ee = EllipticEnvelope(contamination=contamination, random_state=RANDOM_STATE)
ee.fit(Xz)

# 5) Predict and remap to dataset labels (anomaly=1, normal=0).
pred_raw = ee.predict(Xz)               # 1=inlier, -1=outlier
y_pred  = np.where(pred_raw == -1, 1, 0)

# 6) Confusion matrix and short report.
cm = confusion_matrix(y, y_pred, labels=[0, 1])
print("Contamination:", contamination)
print("Confusion matrix [[TN, FP],[FN, TP]]:\n", cm)
print("Classification report:\n",
      classification_report(y, y_pred, target_names=["normal","teardrop"], zero_division=0))

# 7) Plot to match the style of your figure.
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt="d", cmap="YlGnBu", cbar=True)
plt.xlabel("Predicted label")
plt.ylabel("True label")
plt.title("Figure 2.8 – Confusion matrix for elliptic envelope")
out_path = ensure_dir(FIGURES) / "week3_confusion_matrix_elliptic_envelope.png"
plt.tight_layout()
plt.savefig(out_path, dpi=150)
plt.show()
print("Saved confusion matrix to:", out_path)
```
