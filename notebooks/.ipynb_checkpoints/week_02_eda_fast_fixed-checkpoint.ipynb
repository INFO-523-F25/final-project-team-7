```python
# Distributions. (Plot histogram + KDE for selected numeric features on a stratified sample.)
import numpy as np
import pandas as pd
from src.utils import Paths, set_global_seed
from src.io import load_raw_nsl_kdd, map_attack_family
from src.plots import dist_plots

paths = Paths().ensure()
set_global_seed(42)
raw_path = paths.data_raw / 'NSL-KDD.raw'
df = map_attack_family(load_raw_nsl_kdd(raw_path))

# Take a stratified sample to speed up plotting.
sample_per_family = 3000  # adjust down if still slow
df_sample = df.groupby('family', group_keys=False).apply(
    lambda x: x.sample(min(sample_per_family, len(x)), random_state=42)
)

# Choose a small set of representative numeric features.
numeric_cols = [c for c in df_sample.columns if c not in ['protocol_type','service','flag','label','family']]
sel = numeric_cols[:6]  # only first 6 to keep runtime sane

dist_plots(df_sample, sel, paths.figs)
print('Saved distribution plots for columns:', sel)
print('Figures directory:', paths.figs)
```

```python
# Correlations. (Compute Pearson and Spearman heatmaps on numeric features.)
import numpy as np
import pandas as pd
from src.utils import Paths, set_global_seed
from src.io import load_raw_nsl_kdd, map_attack_family
from src.plots import corr_heatmap

paths = Paths().ensure()
set_global_seed(42)
raw_path = paths.data_raw / 'NSL-KDD.raw'
df = map_attack_family(load_raw_nsl_kdd(raw_path))

numeric_cols = [c for c in df.columns if c not in ['protocol_type','service','flag','label','family']]
corr_cols = numeric_cols[:20]
corr_heatmap(df, corr_cols, paths.figs / 'eda_corr_pearson.png', method='pearson')
corr_heatmap(df, corr_cols, paths.figs / 'eda_corr_spearman.png', method='spearman')
print('Saved correlation heatmaps to', paths.figs)
```

```python
# Categorical frequencies. (Top-N bars for protocol_type, service, flag.)
topn_bar(df['protocol_type'], 10, paths.figs / 'eda_top_protocol.png', 'Top Protocol Types')
topn_bar(df['service'], 20, paths.figs / 'eda_top_service.png', 'Top Services')
topn_bar(df['flag'], 10, paths.figs / 'eda_top_flag.png', 'Top Flags')
print('Saved categorical frequency plots to', paths.figs)
```

```python
# Family imbalance. (Report counts and percents.)
fam_counts = df['family'].value_counts()
fam_perc = (fam_counts / len(df) * 100).round(2)
display(pd.DataFrame({'count': fam_counts, 'percent': fam_perc}))
```

```python
# PCA visualization. (2D scatter colored by family.)
num_df = df.select_dtypes(include=[np.number])
X = num_df.fillna(0.0).to_numpy()
pca2 = PCA(n_components=2, random_state=42).fit_transform(X)
fig, ax = plt.subplots()
sns.scatterplot(x=pca2[:,0], y=pca2[:,1], hue=df['family'], ax=ax, s=10, linewidth=0)
ax.set_title('PCA (2D) by Family')
fig.savefig(paths.figs / 'eda_pca2.png', dpi=150, bbox_inches='tight')
plt.close(fig)
print('Saved PCA 2D to', paths.figs)
```

```python
# t-SNE and UMAP. (Use stratified sample for speed.)
sample = df.groupby('family', group_keys=False).apply(lambda x: x.sample(min(1500, len(x)), random_state=42))
X_s = sample.select_dtypes(include=[np.number]).fillna(0.0).to_numpy()
ts = TSNE(n_components=2, random_state=42, perplexity=30, init='pca')
ts2 = ts.fit_transform(X_s)
fig, ax = plt.subplots()
sns.scatterplot(x=ts2[:,0], y=ts2[:,1], hue=sample['family'], s=8, linewidth=0, ax=ax)
ax.set_title('t-SNE (2D) by Family — stratified sample')
fig.savefig(paths.figs / 'eda_tsne.png', dpi=150, bbox_inches='tight')
plt.close(fig)
if umap is not None:
    reducer = umap.UMAP(n_components=2, random_state=42)
    um2 = reducer.fit_transform(X_s)
    fig, ax = plt.subplots()
    sns.scatterplot(x=um2[:,0], y=um2[:,1], hue=sample['family'], s=8, linewidth=0, ax=ax)
    ax.set_title('UMAP (2D) by Family — stratified sample')
    fig.savefig(paths.figs / 'eda_umap.png', dpi=150, bbox_inches='tight')
    plt.close(fig)
print('Saved t-SNE/UMAP plots to', paths.figs)
```

### Notes
- Heavy-tailed numeric features may benefit from log-scale in later analysis.
- Rare classes (R2L, U2R) present risk for model bias; consider class-weighting/SMOTE in Week 4.
